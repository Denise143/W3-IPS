# Importing the libraries





```
#Importing the pandas and numpy libraries which will help in amnipulating the data
import pandas as pd
import numpy as np
```

# Loading Our Tables

```
# We load the Telcom table by using the inbuilt fuction 'pd.read_csv' because it is a csv file
# Some columns have to be renamed to match the other datasets to make merging easier
# The head() is used to display on the five top rows

Telcom1 = pd.read_csv('Telcom_dataset.csv')
Telcom1.rename({'PRODUTC': 'PRODUCT'},axis=1, inplace=True)
Telcom1.head()
```

```
# We load the table and rename columns to match the other datasets
# to display data we have to call on the dataframe

Telcom2 = pd.read_csv('Telcom_dataset2.csv')
Telcom2.rename({'DW_A_NUMBER': 'DW_A_NUMBER_INT', 'DW_B_NUMBER': 'DW_B_NUMBER_INT', 'DATE_TIME': 'DATETIME'},axis=1, inplace=True)
Telcom2
```

```
# The tail() function is used to display the bottom rows
Telcom3 = pd.read_csv('Telcom_dataset3.csv')
Telcom3.rename({'DATE_TIME': 'DATETIME', 'CELLID': 'CELL_ID', 'SIET_ID': 'SITE_ID'},axis=1, inplace=True)
Telcom3.tail(15)
```

```
# This dataset contains the regions that the cellphones were used

Geography = pd.read_csv('cells_geo.csv', delimiter=";")
Geography.rename({'SITE_CODE': 'SITE_ID'}, axis=1, inplace=True)
Geography['SITE_ID'].fillna('DGBD76')
Geography
```

```
table_description = pd.read_excel('cells_geo_description.xlsx')
table_description
```

# Merging and Cleaning


```
# Merging helps give an overview of all the tables.
Frames = [Telcom1, Telcom2, Telcom3]

all_df = pd.concat((Frames), ignore_index=True)

all_df
```

```
# Checking to see whether there are duplicated rows
# The result gives a boolean result of which rows have duplicates
all_df.duplicated()

#Dropping the duplicates
all_df.drop_duplicates()
```

```
#Dropping columns that are not useful
all_df.drop(['CELL_ON_SITE', 'COUNTRY_A', 'COUNTRY_B', 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT'], axis=1, inplace=True )
all_df
```

```
#Merging geography with all_df
Geo_all = all_df.merge(Geography, on='SITE_ID', how='outer')
New_all = Geo_all.dropna(thresh=15)
New_all
```

```
# Dropping duplicates
New_all.duplicated()
New_all.drop_duplicates()
```

```
df = New_all[New_all['VALUE']>0]
df
```
